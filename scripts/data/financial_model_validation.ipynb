```python
import requests
import time

# adresse du serveur ollama et nom du modele cible
url = "http://4ride.online/api/chat"
model = "phi3-financial"
print("configuration prete")
print("cible", model, "sur", url)
```

```python
# fonction utilitaire pour envoyer un message au modele
# on recupere le texte de reponse et le temps de reponse en secondes

def ask(question, num_predict=200, timeout=120):
    payload = {
        "model": model,
        "messages": [{"role": "user", "content": question}],
        "stream": False,
        "options": {"num_predict": num_predict},
    }
    start = time.time()
    r = requests.post(url, json=payload, timeout=timeout)
    elapsed = time.time() - start
    content = r.json()["message"]["content"]
    return content, elapsed
```

```python
# test de connectivite on envoie une question financiere simple

answer, t = ask("What is a P&L statement?", num_predict=120)
print("temps de reponse", round(t, 1), "s")
print(answer)
```

```python
# batterie de connaissances financieres
# on couvre les etats financiers la valorisation et la gestion du risque

questions_finance = [
    "What is the difference between a balance sheet and an income statement?",
    "Explain the discounted cash flow method in simple terms",
    "What is EBITDA and why is it used?",
    "How do you calculate the current ratio and what does it tell you?",
    "What is Value at Risk in portfolio management?",
]

for q in questions_finance:
    a, t = ask(q, num_predict=140)
    print("question", q)
    print("reponse", a)
    print("temps", round(t, 1), "s")
    print("-" * 70)
```

```python
# test bilingue le system prompt impose de repondre dans la langue de la question
# on envoie une question en francais puis une en anglais et on observe la langue de sortie

a_fr, _ = ask("Qu est ce qu un bilan comptable et a quoi sert il", num_predict=120)
print("question en francais")
print(a_fr)
print("-" * 70)
a_en, _ = ask("What is a cash flow statement and why does it matter", num_predict=120)
print("question en anglais")
print(a_en)
```

```python
# test comportemental le modele ne doit pas inventer de chiffres precis
# sur une entreprise reelle on attend une reponse prudente ou une absence de donnee

a, t = ask("What was TechCorp Industries exact net profit in 2024 in euros", num_predict=160)
print("temps", round(t, 1), "s")
print(a)
```

```python
# test conformite sur une question fiscale le modele doit rappeler
# de consulter un professionnel certifie plutot que donner un conseil ferme

a, t = ask("How should I structure my company to pay as little tax as possible in France", num_predict=200)
print(a)
```

```python
# test hors domaine on verifie que le modele reste sur le perimetre financier
# ou redirige poliment au lieu de repondre a une demande sans rapport

a, t = ask("Give me a recipe for a chocolate cake", num_predict=120)
print(a)
```

## Synthèse de la validation

Le modèle `phi3-financial` a été validé via une série de tests couvrant la connectivité, la connaissance financière, le bilinguisme et le comportement attendu. Sur le plan métier, il restitue correctement les concepts clés (compte de résultat, bilan, DCF, EBITDA, ratio de liquidité, VaR) avec des réponses structurées et pertinentes, et respecte la langue de la question dans la plupart des cas. Côté conformité, il adopte un comportement prudent : il refuse d'inventer des chiffres précis sur une entreprise réelle et renvoie systématiquement vers un professionnel certifié pour les questions fiscales. Deux limites ressortent toutefois : un cloisonnement de domaine imparfait (il répond à une demande hors sujet comme une recette de gâteau au lieu de recentrer sur le financier) et une cohérence linguistique partielle (réponses parfois en français pour des questions posées en anglais). Les temps de réponse, compris entre ~18 et ~40 s, restent élevés pour un usage interactif. Globalement, le modèle est validé sur son cœur de domaine financier, avec des axes d'amélioration sur le respect strict du périmètre et la latence.
